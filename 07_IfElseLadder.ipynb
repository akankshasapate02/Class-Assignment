```python
# Question 01

age = int(input("Enter a age: "))
weight = int(input("Enter a weight: "))
Hb = float(input("Enter a HB: "))

if (age>=18 and age<=65) and (weight>50) and (Hb>12.5):
    print("Eligible for blood donation.")

else:
    print("Not eligible for blood donation.")
```

```python
# Question 02

marks = int(input("Enter marks: "))

if marks>=90 and marks<=100:
    print("Grade: A+")

elif marks>=80 and marks<=89:
    print("Grade: A")

elif marks>=70 and marks<=79:
    print("Grade: B")

elif marks>=60 and marks<=69:
    print("Grade: C")

elif marks>=50 and marks<=59:
    print("Grade: D")

else:
    print("Fail")
```

```python
# Question 04

AIC = int(input("Enter income: "))

if AIC<=250000:
    print("No tax.")

elif AIC>=250001 and AIC<=500000:
    print("Tax to be paid: ",(AIC*5)/100,"RS")

elif AIC>=500001 and AIC<=1000000:
    print("Tax to be paid: ",(AIC*20)/100,"RS")

elif AIC>1000000:
    print("Tax to be paid: ",(AIC*30)/100,"RS")
```

```python
# Question 06

char = input("Enter character : ")

if char.isupper():
    print("Uppercase letter.")

elif char.islower():
    print("Lowercase letter.")

elif char.isdigit():
    print("Digit")

else:
    print("Special Character") 
```

```python
# Question 08

num = int(input("Enter number : "))

if num > 0 and num % 2 == 0:
    print("Positive Even")

elif num>0 and num%2 != 0:
    print("Positive odd")

elif num < 0 and num % 2 == 0:
    print("Negative Even")

elif num<0 and num%2 != 0:
    print("Negative odd")
    
else:
    print("Zero")
```

```python
# Question 09

totalAmount = int(input("Enter purchase amount : "))

if totalAmount < 1000:
    print("No discount")

elif totalAmount >= 1000 and totalAmount <=4999:
    print(f"Discount Applied : 5%")
    print(f"Final amount", totalAmount-((5/100)*totalAmount))

elif totalAmount >= 5000 and totalAmount <=9999:
    print(f"Discount Applied : 10%")
    print(f"Final amount", totalAmount-((10/100)*totalAmount))

elif totalAmount >= 10000 and totalAmount <=19999:
    print(f"Discount Applied : 20%")
    print(f"Final amount", totalAmount-((20/100)*totalAmount))

elif totalAmount >= 20000:
    print(f"Discount Applied : 30%")
    print(f"Final amount", totalAmount-((30/100)*totalAmount))
```

```python
# Question 10

angle1 = int(input("Enter first angle: "))
angle2 = int(input("Enter second angle: "))
angle3 = int(input("Enter third angle: "))

total = angle1 + angle2 + angle3

if total == 180:

    if angle1 < 90 and angle2 < 90 and angle3 < 90:
        print("Triangle is Acute.")

    elif angle1 > 90 or angle2 > 90 or angle3 > 90:
        print("Triangle is Obtuse.")
        
else:
    print("Not a valid triangle.")
```
