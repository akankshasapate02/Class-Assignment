```python
#Question 01

def table():
    num = int(input("Enter a number: "))

    for x in range(1,11):
        print(num * x)

table()
```

```python
#Question 02

num = int(input("Enter a number : "))


def evenOdd(num):

    if num%2 == 0:
        return "Even Numebr."
    else :
        return "Odd number."
    
print(evenOdd(num))
```

```python
# Question 03

age = int(input("Enter age : "))

def voting(age):

    if age >= 18:
        return "Eligible for voting."
    else:
        return "Not eligible for voting"

print(voting(age))
```

```python
# Question 04

num1 = int(input("Enter num1 : "))
num2 = int(input("Enter num2 : "))

def arithOperation(num1,num2):

    print(num1+num2)
    print(num1-num2)
    print(num1*num2)
    print(num1/num2)
    print(num1%num2)
    print(num1//num2)

arithOperation(num1,num2)
```

```python
# Question 05

ch = input("Enter Character : ")

def findVowel(ch):

    if ch == 'a' or ch == 'e'or ch == 'i' or ch == 'o' or ch == 'u' or ch == 'A' or ch == 'E'or ch == 'I' or ch == 'O' or ch == 'U':
        return "Vowel Character"
    else :
        return "Consonant Character"

print(findVowel(ch))
```
