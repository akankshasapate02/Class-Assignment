```python
# Question 01

num = int(input("Enter a number"))

if num%4==0 and num%5==0:
    print(f"{num} is divisibe by 4 and 5.")

else:
    print(f"{num} is not divisibe by 4 and 5.")
```

```python
# Question 02

angle1 = int(input("Enter a angle1 value: "))
angle2 = int(input("Enter a angle2 value: "))
angle3 = int(input("Enter a angle3 value: "))

if (angle1+angle2+angle3) == 180:
    print("It is a right angle triangle.")

else:
    print("It is not a right angle triangle.")
```

```python
# Question 03

num1 = int(input("Enter a num1: "))
num2 = int(input("Enter a num2: "))

sum = num1+num2

if sum % 2 == 0:
    print(f"{sum} is Even Number.")

else:
    print(f"{sum} is Odd Number.")
```
