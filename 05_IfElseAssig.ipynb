```python
#Question 05

num1 = int(input("Enter num1: "))

match num1:
    case 0:
        print("Monday")
    case 1:
        print("Tuesday")
    case 2:
        print("Wednesday")
    case 3:
        print("Thursday")
    case 4:
        print("Friday")
    case 5:
        print("Saturday")
    case 6:
        print("Sunday")
    case _:
        print("Wrong Number!!!")
```

```python
# Question 06

char = input("Enter value: ")

if (char>='a' or char<='z') and (char>='A' or char<='Z'):
    print(f"{char} is an alphabet.")

else:
    print(f"{char} is not alphabet.")
```

```python
# Question 07

month = input("Enter Month: ")

if month=="Jan" or month=="March" or month=="May" or month=="July" or month=="Aug" or month=="Oct" or month=="Dec":
    print(f"{month} is a 31-days month.")

elif month=="April" or month=="Jun" or month=="Sep" or month=="Nov":
    print(f"{month} is a 30-days month.")

elif month=="Feb":
    print(f"{month} is a 28-days month.")

else:
    print("Invalid Month name.")
```
