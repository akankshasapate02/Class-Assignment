```python
# Question 2

num = 50

print("Initial value : ", num)
num += 10
print("After += 10 : ", num)
num -= 5
print("After -= 5 : ", num)
num *= 2
print("After *= 2 : ", num)
num /= 4
print("After /= 4 : ", num)
num %= 5
print("After %= 5 : ", num)
num //= 3
print("After //= 3 : ", num)
```

```python
#Question 5

a = 100
b = 100
list1 = [1, 2, 3]
list2 = [1, 2, 3]

#Integer Variables : 
print("a = ", a , end = " ")
print("b = ", b)
print("id(a) = ", id(a))
print("id(b) = ", id(b))
print("a is b = ", a is b)
print("a is not b = ", a is not b ,"\n")

#List Variables :
print("list1 = ", list1 , end = " ")
print("list2 = ", list2) 
print("id(list1) = ", id(list1))
print("id(list2) = ", id(list2))
print("list1 is list2 = ", list1 is list2)
print("list1 == list2 = ", list1 == list2)
print("list1 is not list2 = ", list1 is not list2)
```

```python
#Question 7

x = 12
y = 10

print("x = ", x)
print("y = ", y)
print("x & y (Bitwise AND) : ", x&y)
print("x | y (Bitwise OR) : ", x|y)
print("x ^ y (Bitwise XOR) : ", x^y)
print("~x (Bitwise NOT) : ", ~x)
print("x << 2 (left shift by 2) : ", x<<2)
print("x >> 2 (right shift by 2) : ", x>>2)
```

```python
#Question 8

intVal = 10
floatVal = 5.5
stringVal = "Python"

print("intVal = 10 Type : ", type(intVal))
print("floatVal = 5.5 Type : ", type(floatVal))
print("stringVal = 'Python' Type : ", type(stringVal))

print("intVal + floatVal  = ", intVal + floatVal , " : " , type(intVal + floatVal))
print("intVal * floatVal  = ", intVal * floatVal , " : " , type(intVal * floatVal))
print("floatVal / intVal  = ", floatVal / intVal , " : " , type(floatVal / intVal))
print("stringVal *2 = ", stringVal *2 , " : ", type(stringVal))
```
