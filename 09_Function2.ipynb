```python
# Question 01

num1 = int(input("Enter number num1 : "))
num2 = int(input("Enter number num2 :"))
num3 = int(input("Enter number num3 : "))

def max(a,b,c):
    
    if a>b and a>c:
        print(f"{a} is a maximum number.")
    elif b>a and b>c:
        print(f"{b} is a maximum number.")
    else:
        print(f"{c} ia a maximum number.")

max(num1,num2,num3)
```
